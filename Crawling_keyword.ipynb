```python
import requests
from bs4 import BeautifulSoup
# 한양대학교 창업지원단에 대한 웹 크롤링
# 크롤링 동작 확인
response = requests.get("https://startup.hanyang.ac.kr")
soup = BeautifulSoup(response.text, "html.parser")
print(soup)
```

```python
# 웹 사이트의 모든 text추출
soup.get_text()
```

```python
# 홈페이지 링크 추출(a href)
a_tag = soup.select("a")
a_tag
```

```python
for i in a_tag:
    try:
        link = i.attrs["href"]
        print(link)
    except:
        pass
```

```python
# "https://", "http://"는 외부웹사이트이고, "/", "#"홈페이지 주소와 중복되며, "/en"는 영어 이므로 제외처리
for i in a_tag:
    try:
        link = i.attrs["href"]
        if "//" not in link and len(link) != 1 and link != "/en":
            print(link)
    except:
        pass
```

```python
# 추출된 링크에 대한 최종 url링크 생성
for i in a_tag:
    try:
        link = i.attrs["href"]
        if "//" not in link and len(link) != 1 and link != "/en":
            print("https://startup.hanyang.ac.kr" + link)
    except:
        pass
```
