```python
import requests
from bs4 import BeautifulSoup
import re
import json

response = requests.get("https://startup.hanyang.ac.kr/board/media_report/list")
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
target = soup.select("div.container")
target  # 위의 url은 자료실이고, 페이지별로 구성되어있음, 출력결과 비동기방식으로 확인됨
```

```python
word = []
for i in target:
        k = str(i).split("</")
        for j in k:
            keyword = re.findall("(?<=\>)[A-Z a-z 0-9 가-힣 /()!@#$%&*\-~+=?.,:;·]+", j)
            if keyword:
                for t in keyword:
                    if t not in word:
                        word.append(t)
word # 이 부분이 빈배열이면 비동기, 그렇지 않으면 동기 방식임
```
